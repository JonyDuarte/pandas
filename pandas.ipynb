## Passo 2
---
Após a instalação do **Pandas**, iremos importar a biblioteca e abrir o **dataset** no *notebook* para a melhor exibição de seus dados. Mas antes de digitar o código Python, altere o **kernel** do *notebook* para o ambiente virtual Python criado junt com o projeto.

Após isso, crie a célula de código, mas quando executar a célula, o **Jupyter** irá pedir a instalçao do pacote do ***ipykernel***. Prossiga com a instalçao para a visualização do resultado do código-fonte. 

Segue a célula do código abaixo:

```python
# importação da biblioteca
import pandas as pd

# lê os dados da tabela
tabela = pd.read_csv("cancelamentos.csv")

# mostra dos dados da tabela no notebook
display(tabela)
```

## Passo 3
---

Depois de puxar os dados da tabela para o notebook, faça o tratamento dos dados. Obeserve, por exemplo, que algumas colunas, como a coluna do ID é completamente inútil para a nossa análise. Portanto, iremos fazer outra célula de código para reirar essa coluna da nossa análise> Observe a célula de código abaixo:

```python
tabela = tabela.drop("CustomerID", axis=1)
display(tabela)
```

A análise anterior mostra que temos um total de **881666** linhas. Entretanto, a saída não mostra quantas dessas linhas estão com espaços vazios, que não poderão ser considerados pela futura análise.

Dessa forna, precisaremos contabilizar o número de linhas com algum tipo de dado, e quantas linhas possuem valor nulo.

```python
display(tabela.info())
```

O resultado acima mostra a quantidade de linhas de cada campo, contando inclusive com as linhas vazias. Repare que o número de linhas é diferente em alumas colunas. Na próxima célula de código, precisaremos eliminar as células com valor nulo:

```python
tabela = tabela.dropna()
display(tabela.info())
```

Observe agora que o número de linhas é igual para todas as colunas, que por sua vez, são iguais ao número total de linhas da tabela inteira. Agora que fizemos o **tratamento de dados**, vamos para o mais importante: a **Análise de Dados**. 

## Passo 4
---

Agora finalmente iremos iniciar a análise de dados propriamente dita. Começaremos analisando a taxa de cancelamento dos planos de assinatura da empresa. Verificaremos qual é a taxa de cancelament e de onde ela vem:

```python
display(tabela["cancelou"].value_counts())
```

Temos um número. A linha com valores **1.0** representam os cancelamentos. As linhas **0.0** representam os não cancelamentos. Mas para um leigo, dica difícil dizer se os valores são bons ou ruin. Portanto, vamos transformar esses números em porcentagem.

```python
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

O `.map()` atribui a lista uma operação que transforma os valores em **porcentagem**.

Veja que, de acordo com a nossa análise, mais da metade dos clientes cancelaram o serviço. O próximo passo é tentar entender de onde vem uma taxa tão alta de cancelamento.

## Passo 5
---
Precisaremos analisar por partes. Vamos começar pela duração dos contratos. No *dataset*, existem 3 tipos de contratos: **Anual**, **Trimestral** e **Mensal**. Vejamos qual a duração de cada um dos tipos de contratos:

```python
display(tabela["duracao_contrato"].value_counts())
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.2%}".format))
```

Observamos algo interessante: os contratos anuais e trimestrais tem uma divisão quase igual, mas há uma discrepância muito maior na duração dos contratos mensais.

Vamos agrupar as informações da coluna **duracao_contrato** e depois fazer a média das informações. Para isso, iremos usar a função `groupby()` e repassar o nome da coluna nela:

```python
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
```

Aqui temos uma informação alarmante: os contratos do tipo *Mensal* tiveram exatos **100 % DE CANCELAMENTO!**
